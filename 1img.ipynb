```python
from tensorflow.keras.preprocessing import image
```

```python
img = image.load_img('/content/drive/MyDrive/Dataset/1029549.jpg',target_size=(224,224))
```

```python
img_array = image.img_to_array(img)
```

```python
img_array
```

```python
img_array.shape
```

```python
import numpy as np
expanded_img_array = np.expand_dims(img_array,axis=0)
print(expanded_img_array.shape)
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input
import tensorflow
from tensorflow.keras.preprocessing import image                                     # converting images to arrays, resizing, etc.
from tensorflow.keras.layers import GlobalMaxPooling2D
```

```python
model = ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
model.trainable = False
model = tensorflow.keras.Sequential([model,GlobalMaxPooling2D()])
```

```python
preprocessed_img = preprocess_input(expanded_img_array)
print(preprocessed_img.shape)
preprocessed_img
```

```python
model.predict(preprocessed_img)
```

```python
model.predict(preprocessed_img).flatten()
```

```python
model.predict(preprocessed_img).flatten()/norm(model.predict(preprocessed_img).flatten())
```

```python
from numpy.linalg import norm
norm(model.predict(preprocessed_img).flatten())
```
